```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").appName("my app").enableHiveSupport().getOrCreate()
spark.conf.set("spark.sql.repl.eagerEval.enabled", True) # Property used to format output tables better
spark
```

```python
!wget https://jacobceles.github.io/knowledge_repo/colab_and_pyspark/cars.csv
```

```python
# Load data from csv to a dataframe. 
# header=True means the first row is a header 
# sep=';' means the column are seperated using ''
df = spark.read.csv('cars.csv', header=True, sep=";")
df.show(5)
```

```python
sc = spark.sparkContext
from pyspark.sql import HiveContext
# hiveCtx = HiveContext(sc)
sqlContext = HiveContext(sc)
```

```python
df.createOrReplaceTempView("temp")
# Select all data from temp table
spark.sql("select * from temp limit 5").show()
# Select count of data in table
spark.sql("select count(*) as total_count from temp").show()
```

```python
spark.sql("create table test as select * from temp limit 5")
```

```python
spark.sql(" select * from test limit 5")
```

```python
sqlContext.sql("""drop table if exists test2""")
sqlContext.sql("create table test2 as select * from temp")
```

```python
sqlContext.sql("select * from test2")
```

```python
sqlContext.sql("desc formatted test2").show(20, False)
```
